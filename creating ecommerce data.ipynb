```python
from faker import Faker
fake = Faker()
```

```python
import pandas as pd
customer_code = []
for _ in range(1000):
    fakecode = fake.postcode()
    cust_code = 'CUST' + fakecode
    customer_code.append(cust_code)
    
ecommerce = pd.Series(customer_code, name='customer_id')
```

```python
product_code = []
for _ in range(30):
    fakecode2 = fake.postcode()
    prod = 'PROD' + fakecode2
    product_code.append(prod)
    
product_id = pd.Series(product_code, name='product_id')
ecommerce = pd.concat([ecommerce, product_id], axis=1)
```

```python
order_code = []
for _ in range(1000):
    fakecode = fake.postcode()
    ord_code = 'ORDR' + fakecode
    order_code.append(ord_code)
    
ecommerce['order_id'] = pd.Series(order_code, name='order_id')
```

```python
order_item_code = []
for _ in range(1000):
    fakecode = fake.postcode()
    ord_itm_code = ['ORDR00' + fakecode + '-0' + str(i) for i in range(1, 1001)]
    for x in ord_itm_code:
            order_item_code.append(x)
ecommerce['order_item_id'] = pd.Series(order_item_code, name='order_item_id')
```

```python
customer_name = []
for _ in range(1000):
    fakename = fake.name()
    customer_name.append(fakename)
   
ecommerce['customer_name'] = pd.Series(customer_name, name='customer_name')
```

```python
date_code = []
for _ in range(1000):
    fakedate = fake.date_between(start_date='-2y', end_date='today')
    date_code.append(fakedate)

ecommerce['date'] = pd.Series(date_code, name='date')
```

```python
import random

products = [
    "FreePrints 4x6 Photo Print",
    "FreePrints Square Photo Print",
    "FreePrints Glossy Photo Print",
    "FreePrints Matte Photo Print",
    "FreePrints 8x10 Photo Print",
    "FreePrints Hardcover Photobook",
    "FreePrints Softcover Photobook",
    "FreePrints Layflat Photobook",
    "FreePrints Mini Photobook",
    "FreePrints Wall Calendar",
    "Personalized Holiday Greeting Card",
    "Custom Wedding Invitation Card",
    "Birth Announcement Card",
    "Thank You Card with Photo",
    "Foil-Pressed Greeting Card",
    "Personalized Business Card",
    "Personalized Photo Mug",
    "Custom Name Necklace",
    "Personalized Photo Blanket",
    "Engraved Wooden Keepsake Box",
    "Personalized Cutting Board",
    "Custom Photo Pillow",
    "Custom Printed T-Shirt",
    "Personalized Hoodie",
    "Custom Coffee Mug",
    "Personalized Phone Case",
    "Custom Tote Bag",
    "Personalized Children’s Storybook",
    "Custom Name Puzzle Book",
    "Personalized Birthday Storybook"
]

# ------------------------------
# Create Product Dimension DataFrame
# ------------------------------
# product_dim stores the mapping of product IDs to product names
# - 'product_id': list of unique product IDs (generated earlier)
# - 'product_name': corresponding list of product names
product_dim = pd.DataFrame({
    'product_id': product_code,          # your generated product_ids
    'product_name': products             # the 30 product names
})

# ------------------------------
# Repeat product IDs to match ecommerce dataframe length
# ------------------------------
# ecommerce has more rows than unique products (e.g., 1000 rows vs 30 products)
# repeats = number of times to repeat product list to cover all rows
# len(ecommerce) // 30 → number of full sets of 30 products
# +1 → ensures extra IDs to cover remainder rows
repeats = len(ecommerce) // 30 + 1

# product_code * repeats → repeats the product list 'repeats' times
# [:len(ecommerce)] → slice to ensure exactly one product ID per row
product_ids_rows = (product_code * repeats)[:len(ecommerce)]

# ------------------------------
# Shuffle product IDs randomly
# ------------------------------
# random.shuffle ensures product distribution is random across rows
random.shuffle(product_ids_rows)

# ------------------------------
# Assign shuffled product IDs to ecommerce dataframe
# ------------------------------
# Each row in ecommerce now has a product_id
ecommerce['product_id'] = product_ids_rows

ecommerce = ecommerce.merge(product_dim, on='product_id', how='left')
```

```python
def random_spend(platform):
    if platform == 'Google Ads':
        return round(random.uniform(40, 120), 2)
    elif platform == 'Facebook Ads':
        return round(random.uniform(25, 100), 2)
    elif platform == 'Instagram Ads':
        return round(random.uniform(20, 90), 2)
    elif platform == 'Mailchimp':
        return round(random.uniform(5, 30), 2)
    else:
        return 0 

ecommerce['spend'] = ecommerce['platform_name'].apply(random_spend)
```
